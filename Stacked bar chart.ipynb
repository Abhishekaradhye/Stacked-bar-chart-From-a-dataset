```python
import pandas as pd
from matplotlib import pyplot as plt

df1 = pd.read_csv('/content/Batsmen.csv')
df = pd.DataFrame(df1)

df
```

```python
index1 = {0:'V. Kohli',1:'R. Sharma',2:'S. Dhawan',3:'M. Dhoni',4:'S. Iyer'}

df.rename(index=index1,inplace = True)
df
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(7, 4))

ax.bar(df.index, df["0s"], label="0s")
ax.bar(df.index, df["1s"], bottom=df["0s"], label="1s")
ax.bar(df.index, df["2s"], bottom=df["0s"] + df["1s"], label="2s")
ax.bar(df.index, df["3s"], bottom=df["0s"] + df["1s"] + df["2s"], label="3s")
ax.bar(df.index, df["4s"], bottom=df["0s"] + df["1s"] + df["2s"] + df["3s"], label="4s")
ax.bar(df.index, df["6s"], bottom=df["0s"] + df["1s"] + df["2s"] + df["3s"] + df["4s"], label="6s")

for idx in df.index:
    ab = 0
    for col in df.columns:
        y = df.loc[idx, col]
        value = df.loc[idx, col]
        total = df.loc[idx, :].sum()
        ax.text(x=idx, y=ab + (y / 2), s=f"{round(100 * value / total, 1)}%", fontsize=8,
            ha="center",color="w")
        ab += y

plt.xticks(df.index, df.index)
ax.legend()
plt.title("Deliveries tackled by batsmen in one innings")

plt.show()
```
